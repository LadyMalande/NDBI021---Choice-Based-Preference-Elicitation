```python
from matrix_factorization import BaselineModel, KernelMF, train_update_test_split
from itertools import repeat
from scipy.spatial.distance import cdist
from scipy.spatial import KDTree
from scipy.spatial import cKDTree
import numpy as np 
from sklearn.neighbors import NearestNeighbors
import random

import pandas as pd
from sklearn.metrics import mean_squared_error

# Movie data found here https://grouplens.org/datasets/movielens/
cols = ["user_id", "item_id", "rating", "timestamp"]
movie_data = pd.read_csv(
    "ratings.csv", names=cols, sep=";", usecols=[0, 1, 2], engine="python"
)

cols2 = ["item_id", "item_name"]
movie_names = pd.read_csv(
    "movies.csv", names=cols2, sep=";", usecols=[0, 1], engine="python"
)



# Declare a list that is to be converted into a column
#movie_name = []

#movie_data = pd.reset_index()  # make sure indexes pair with number of rows
#for index, row in movie_data.iterrows():
#    for index2, row2 in movie_names.iterrows():
#        if row['item_id'] == row2['item_id']:
#            movie_name.append(row2['item_name'])
#            print(row['item_id'], row2['item_name'])
            
df_outer = pd.merge(movie_data, movie_names, on='item_id', how='outer') #here id is common column
print(len(df_outer))
df_outer = df_outer.dropna()
print(len(df_outer))
df_outer2 = df_outer.copy(deep=False)
print(df_outer.loc[df_outer.rating == 'NaN','item_name'])

X = df_outer[["user_id", "item_name"]]
y = df_outer["rating"]

#df_outer
    
# Adding the new column to the movie_data
#movie_data['movie_name'] = movie_name
 
# Observe the result


# Split data for learning
(
    X_train_initial,
    y_train_initial,
    X_train_update,
    y_train_update,
    X_test_update,
    y_test_update,
) = train_update_test_split(df_outer, frac_new_users=0.2)

# Initial training
matrix_fact = KernelMF(n_epochs=20, n_factors=20, verbose=1, lr=0.001, reg=0.005)
matrix_fact.fit(X_train_initial, y_train_initial)

# Update model with new users
matrix_fact.update_users(
    X_train_update, y_train_update, lr=0.001, n_epochs=20, verbose=1
)
pred = matrix_fact.predict(X_test_update)
rmse = mean_squared_error(y_test_update, pred, squared=False)
# print(f"\nTest RMSE: {rmse:.4f}")

# Get recommendations
user = 189
items_known = X_train_initial.query("user_id == @user")["item_id"]

matrix_fact.recommend(user=user, items_known=items_known)



```

```python
ourFactor = []

for factorId in range(5):
    # Choose only the nth factor from the item features
    factor1 = matrix_fact.item_features.flatten()[factorId::20].copy()
    print(factor1.enumerate())
    
    # Sort the factor andremember the movie id with it
    sortedFactor1 = sorted(enumerate(factor1), key=lambda i: i[1])

    #Get subarrays from the appropriate area: 20-35% and 65-80%
    partitionA = sortedFactor1[1117:3006]
    partitionB = sortedFactor1[5582:6871]

    #print(partitionA)

    # Get only movie indices, omit the factors
    itemIndexFromA = [i[0] for i in partitionA]
    itemIndexFromB = [i[0] for i in partitionB]

    #print(set(itemIndexFromA) & set(populars))

    # Get matches: Items that are in correct range and are popular 
    popularFromPartitionA = set(itemIndexFromA) & set(populars)
    popularFromPartitionB = set(itemIndexFromB) & set(populars)
    #print(popularFromPartitionB)

    #For partition A and B, choose 5 movie samples from popular movies. If there are less than 5 popular movies, choose arbitrary
    if 5 > len(popularFromPartitionA):
        sampleForPartitionA = random.sample(popularFromPartitionA, len(popularFromPartitionA)) 
        for i in range(5-len(sampleForPartitionA)):
            oneSampleFromPartitionA = random.sample(itemIndexFromA, 1)
            sampleForPartitionA = sampleForPartitionA + oneSampleFromPartitionA
    else:
        sampleForPartitionA = random.sample(popularFromPartitionA, 5) 

    filteredPartitionA = [i for i in partitionA if i[0] in sampleForPartitionA]
    factorsA = [i[1] for i in filteredPartitionA]
    averageA = sum(factorsA)/5
    #print(averageA)

    if 5 > len(popularFromPartitionB):
        sampleForPartitionB = random.sample(popularFromPartitionB, len(popularFromPartitionB)) 
        for i in range(5-len(sampleForPartitionB)):
            oneSampleFromPartitionB = random.sample(itemIndexFromB, 1)
            sampleForPartitionB = sampleForPartitionB + oneSampleFromPartitionB
    else:
        sampleForPartitionB = random.sample(popularFromPartitionB, 5) 
        
    filteredPartitionB = [i for i in partitionB if i[0] in sampleForPartitionB]
    factorsB = [i[1] for i in filteredPartitionB]
    averageB = sum(factorsB)/5
    #print(averageB)

    #print(len(sampleForPartitionB))
    print("\n")
    print("Sample 1:")
    for i in range(5):
        #print(sampleForPartitionA)
        #print(sampleForPartitionA[i])
        print(movie_names.loc[movie_names.item_id == sampleForPartitionA[i],'item_name'].item())
    
    print("Sample 2:")
    for i in range(5):
        #print(i)
        #print(sampleForPartitionB)
        #print(sampleForPartitionB[i])
        print(movie_names.loc[movie_names.item_id == sampleForPartitionB[i],'item_name'].item())    

    preferredSample = input("Enter number your preferred sample: ")

    toOurFactor = 0
    if(preferredSample == '1'):
        toOurFactor = averageA
    else:
        toOurFactor = averageB
        
    ourFactor.append(toOurFactor)

print(ourFactor)
```
